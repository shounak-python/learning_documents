1. Make virtual env:<br>
`mkvirtualenv django4 --python=/usr/bin/python3.9
pip install django==4.0.7
workon django`

2. To check python verison, enter:<br>

`python --version`<br>
 `>Python 3.9.13`<br><br>
`python -m django --version`<br>
 `>4.0.7`
 
3. To start project, use:<br>
`django-admin startproject project_name`
 
4. To check staus of your Project, use:<br>
`python3 manage.py check`

5. To migrate databases and other files, use:<br>
`python3 manage.py migrate`

6. Create Django App: <br>
`python manage.py startapp authentication`

7. Launch Server: <br>
`python manage.py runserver`
