# General Kafka Commands


1. Download Kafka:
`wget https://archive.apache.org/dist/kafka/2.8.0/kafka_2.12-2.8.0.tgz`

<br>2. Extract Kafka:
`tar -xzf kafka_2.12-2.8.0.tgz`

<br>3. Start Zookeper (open new terminal):
`cd kafka_2.12-2.8.0
bin/zookeeper-server-start.sh config/zookeeper.properties`

<br>4. Start Kafka Broker Service (open new terminal):
`cd kafka_2.12-2.8.0
bin/kafka-server-start.sh config/server.properties`

<br>5. Create Topic (open new terminal):
`cd kafka_2.12-2.8.0
bin/kafka-topics.sh --create --topic news --bootstrap-server localhost:9092`

<br>6. Start Producer (open new terminal):
`cd kafka_2.12-2.8.0
bin/kafka-console-producer.sh --topic news --bootstrap-server localhost:9092`<br>

Once the Producer starts and you get `>` prompt, you can add your messages.
<br>`>Good morning`
<br>`>Good day`
<br>`>Enjoy the Kafka lab`

<br>7. Start Consumer (open new terminal):
`cd kafka_2.12-2.8.0
bin/kafka-console-consumer.sh --topic news --from-beginning --bootstrap-server localhost:9092`
You will get output as the messages you sent via Producer.

<br>8. Folders in kafka:

|Directory	Contents| |
| :---|:--- |
|bin | shell scripts to control kafka and zookeeper|
|config | configuration files|
|logs | log files forkafka and zookeeper|

1. Set up Consumer Group:
`bin/kafka-console-consumer.sh --bootstrap-server localhost:9092 --topic bankbranch --group atm-app`

2. Use key to allocate the proper partition

3. Use describe to see status of consumer group
`bin/kafka-consumer-groups.sh --bootstrap-server localhost:9092 --describe --group atm-app`

4. Reset offsets
`bin/kafka-consumer-groups.sh --bootstrap-server localhost:9092  --topic bankbranch --group atm-app --reset-offsets --to-earliest --execute`

`bin/kafka-consumer-groups.sh --bootstrap-server localhost:9092  --topic bankbranch --group atm-app --reset-offsets --shift-by -2 --execute`
